```python
import pandas_datareader.data as web
import datetime 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = datetime.datetime(2015,1,1)
end   = datetime.datetime(2021,7,21)
```

```python
inf = web.DataReader("INFY.NS",'yahoo',start,end)
tcs = web.DataReader("TCS.NS",'yahoo',start,end)
wipro = web.DataReader("WIPRO.NS",'yahoo',start,end)
```

```python
#Saving data into csv file
inf.to_csv('Infosys.csv')     
tcs.to_csv('Tcs.csv')
wipro.to_csv('Wipro.csv')
```

# ----------------Daily Return Percentage---------------

```python
inf['returns'] = (inf['Close']/inf['Close'].shift(1))-1
tcs['returns'] = (tcs['Close']/inf['Close'].shift(1))-1
wipro['returns'] = (wipro['Close']/inf['Close'].shift(1))-1
```

```python
tcs.head()
```

```python
wipro.head()
```

```python
inf['returns'].hist(bins=100,label='Infosys',figsize=(15,8))
tcs['returns'].hist(bins=100,label='TCS')
wipro['returns'].hist(bins=100,label='WIPRO')
plt.legend()
```

# Cumulative Return

```python
inf['return'] = (1 + inf['returns']).cumprod()
tcs['return'] = (1 + tcs['returns']).cumprod()
wipro['return'] = (1 + wipro['returns']).cumprod()
```
